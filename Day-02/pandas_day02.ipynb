```python
doori = pd.DataFrame(
    data=[['12345', 100, 200, 300], ['23456', 400, 500, 600], ['34567', 700, 800, 900]],
    columns=['zip', 'fact', 'ware', 'ret']
)
doori
```

```python
sns.barplot(data=ds, x='zip', y='dist', hue='type')
```
